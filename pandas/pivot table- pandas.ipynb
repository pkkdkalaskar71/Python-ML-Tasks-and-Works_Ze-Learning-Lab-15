```python
import pandas as pd
import numpy as np
d  = {'First Name': ['Ashok', 'Rana', 'Riya', 'Yogesh', 'Manoj', ],
      'Type': ['Full-time Employee', 'Contract', 'Full-time Employee','Part-time Employee', 'Full-time Employee'],
      'Department': ['Admin', 'Tech', 'Admin','Tech', 'Magmt'],
      'YoE': [1,4,5,3,6],
      'Salary': [20, 50, 10, 10, 20]}
df = pd.DataFrame(d)

df
```

```python
pd.pivot_table(df, index = 'Type',values= 'Salary',columns='Department',)
```

```python
pd.pivot_table(df, index = 'Department',values= 'Salary',aggfunc = ['sum','mean','count'])
```
